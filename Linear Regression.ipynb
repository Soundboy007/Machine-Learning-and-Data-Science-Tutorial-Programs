```python
%matplotlib inline
import numpy as np
from pylab import *
pageSeeds=np.random.normal(3.0,1.0,1000)
purchaseAmount=100-(pageSeeds+np.random.normal(0,0.1,1000))*3
scatter(pageSeeds,purchaseAmount)
```

```python
from scipy import stats
slope,intercept,r_value,p_value,std_err=stats.linregress(pageSeeds,purchaseAmount)
```

```python
print(r_value**2)
print(p_value)
print(std_err)
```

```python
import matplotlib.pyplot as plt
def predict(x):
    return slope*x+intercept
fitline=predict(pageSeeds)
plt.scatter(pageSeeds,purchaseAmount)
plt.plot(pageSeeds,fitline,c='r')
plt.show()
```
